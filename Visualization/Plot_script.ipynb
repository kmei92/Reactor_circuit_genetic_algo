```python
import sys
import graphviz
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Load the performance data 
num_units = [5, 10, 15]
gene_change_rate = 0.01 
crossover_rate = 0.9 
max_steps = 500
performance = []
for units in num_units:
    performance.append(np.loadtxt("./performance_list"+ "_" + str(units) +"_" + str(max_steps) +  "_gcr"+ str(gene_change_rate)+"_cr" + str(crossover_rate) + ".txt"))
```

```python
# Revenue vs steps
fig1 = plt.figure(figsize=(8, 8))
ax1 = plt.subplot(111)

Label = ['5 units','10 units', '15 units']
color_label = ['c.--','m.--', 'k.--']
for id in range(3):
    ax1.plot(performance[id][:,0], color_label[id], label= Label[id] )

ax1.set_xlabel('$steps$', fontsize=16)
ax1.set_ylabel('$Revenues$', fontsize=16)

ax1.set_title('$Revenues$ $with$ $different$ $number$ $of$ $units$ \n(10000 steps, 100 parents, 0.01 mutation rate, 0.9 crossover rate)', fontsize=16)
ax1.grid()
ax1.legend(loc = 'best', fontsize = 14)
```

```python
# Draw the graph
def draw_graph(filename, outpath, id):
    graph = graphviz.Digraph()
    data = np.loadtxt(filename)
    vec = [int(x) for x in data]
    N = (len(vec)-1)//2

    def label(x):
        if x<N:
            return 'Unit %s'%(x)
        elif x==N:
            return 'Concentrate'
        else:
            return 'Tailings'

    graph.attr(rankdir='LR')
    graph.attr('node', shape='rectangle')

    graph.edge('Feed', label(vec[0]), color='blue',
               headport='w', tailport='e',
               arrowhead='normal', arrowtail='normal')

    for _ in range(N):
        graph.edge(label(_), label(vec[2*_+1]), color='blue',
                   headport='w', tailport='n',
                   arrowhead='normal')
        graph.edge(label(_), label(vec[2*_+2]), color='red',
                   headport='w', tailport='s',
                   arrowhead='normal')

    with graph.subgraph() as s:
        s.attr(rank = 'source')
        for n in ('Feed',):
            s.node(n)

    with graph.subgraph() as s:
        s.attr(rank = 'sink')
        for n in ('Tailings', 'Concentrate'):
            s.node(n)

    
    graph.attr(label="Vector: %s \n Revenue: %s"%(vec, performance[id][-1,0]))
    graph.render(outpath, view=True)


for id in range(3):
    filename = "./final_vec"+ "_" + str(num_units[id]) + "_" + str(max_steps) + "_gcr"+ str(gene_change_rate)+"_cr" + str(crossover_rate) + ".txt"
    outpath = "./graph/configuration"+ "_" + str(num_units[id]) + "_" + str(max_steps)+ "_gcr"+ str(gene_change_rate)+"_cr" + str(crossover_rate)
    draw_graph(filename, outpath, id)
```
